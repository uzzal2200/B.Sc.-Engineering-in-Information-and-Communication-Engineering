```python
import numpy as np
import matplotlib.pyplot as plt


n  = np.arange(-10, 11)
alpha = -0.1 + 0.3j

x = np.exp(alpha * n)

plt.figure(figsize=(12 ,10))

plt.subplot(221)
plt.stem(n, np.real(x))
plt.title(" Real Part ")
plt.xlabel(" n ")


plt.subplot(222)

plt.stem(n, np.imag(x))
plt.title(" Imaginary Part ")
plt.xlabel(" n")

plt.subplot(223)
plt.stem(n, np.abs(x))

plt.title(" Magnitude ")

plt.xlabel(" n ")

plt.subplot(224)
plt.stem(n, np.angle(x))
plt.title(" Phase (Degres)")
plt.xlabel('n')
```
