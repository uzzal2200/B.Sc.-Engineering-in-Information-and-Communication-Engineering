```python
# a. x(n)=2δ(n+2)−δ(n−4), −5 ≤ n≤5.

import numpy as np
import matplotlib.pyplot as plt

# ইমপালস ফাংশন ডিফাইন
def impulse(n, shift):
    return np.where(n == shift, 1, 0)

# n এর পরিসীমা: -5 থেকে 5 পর্যন্ত
x = np.arange(-5, 6)

# 2δ(n+2) অংশ
shift = -2
impulse_signal = 2 * impulse(x, shift)

# -δ(n−4) অংশ
shift = 4
impulse_signal2 = impulse(x, shift)

# মোট সংকেত x(n) = 2δ(n+2) − δ(n−4)
subtraction = impulse_signal - impulse_signal2

# গ্রাফ আঁকা
plt.stem(x, subtraction)
plt.title("x(n) = 2δ(n+2) − δ(n−4)")
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()
```

```python
#  b. x(n)=n[u(n)−u(n−10)]+10e−0.3(n−10)[u(n−10)−u(n−20)], 0 ≤ n ≤ 20.

# ইউনিট স্টেপ ফাংশন ডিফাইন
def unit_step(n):
    return np.where(n >= 0, 1, 0)

# n এর পরিসীমা: 0 থেকে 20 পর্যন্ত
n = np.arange(0, 21)

# প্রথম অংশ: n[u(n)−u(n−10)]
x1 = n * (unit_step(n) - unit_step(n-10))

# দ্বিতীয় অংশ: 10e^(−0.3(n−10))[u(n−10)−u(n−20)]
x2 = 10 * np.exp(-0.3 * (n-10)) * (unit_step(n-10) - unit_step(n-20))

# মোট সংকেত x(n) = x1 + x2
x = x1 + x2

# গ্রাফ আঁকা
plt.stem(n, x)
plt.title("x(n) = n[u(n)−u(n−10)]+10e^(−0.3(n−10))[u(n−10)−u(n−20)]")
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()
```

```python
# c. x(n)=cos(0.04πn)+0.2w(n), 0 ≤ n ≤ 50, where w(n) is aGaussian
#  random sequence with zero mean and unit variance

n = np.arange(0, 51)

noise = np.random.normal(0, 1, len(n))

x = np.cos(0.04*np.pi*n) + 0.2*noise;

plt.stem(n,x)
plt.xlabel('n')
plt.ylabel('x(n)')
```
