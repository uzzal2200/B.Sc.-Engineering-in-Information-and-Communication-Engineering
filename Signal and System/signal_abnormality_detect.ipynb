```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
# Generate a sample signal
np.random.seed(42)
time = np.linspace(0,10,500)
signal = np.sin(time) + np.random.normal(0,0.1,len(time))
```

```python
# Introduce abnormalities
signal[100] = 2
signal[300:305] = -2
upper_threshold = 1.5
lower_threshold = -1.5

anomalies = (signal > upper_threshold) | (signal < lower_threshold)
plt.figure(figsize=(12,6))
plt.plot(time, signal, label="signal")
plt.scatter(time[anomalies],signal[anomalies],color='red',label='Anomalies',zorder=5)
plt.axhline(upper_threshold,color='green',linestyle="--",label="upper_threshold")
plt.axhline(lower_threshold,color='orange',linestyle="--",label="lower_threshold")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Signal Abnormality Detection")
plt.legend()
plt.show()

```

Convolution

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal and kernel
signal = np.array([1, 2, 3, 4, 5])
kernel = np.array([1, 0, -1])  # Example of an edge-detection kernel

# Perform convolution using NumPy
convolved_signal = np.convolve(signal, kernel, mode='full')

# Display the result
print("Original Signal:", signal)
print("Kernel:", kernel)
print("Convolved Signal:", convolved_signal)

# Plot the original signal and convolved signal
plt.figure(figsize=(10, 4))

# Plot the original signal
plt.subplot(1, 2, 1)
plt.stem(range(len(signal)), signal, label='Signal')  # Removed `use_line_collection`
plt.title('Original Signal')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Plot the convolved signal
plt.subplot(1, 2, 2)
plt.stem(range(len(convolved_signal)), convolved_signal, label='Convolved Signal', 
         linefmt='r', markerfmt='ro')  # Removed `use_line_collection`
plt.title('Convolved Signal')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal and kernel
signal = np.array([1, 2, 3, 4, 5])
kernel = np.array([1, 0, -1])  # Example of a kernel for correlation

# Perform correlation using NumPy
correlated_signal = np.correlate(signal, kernel, mode='full')

# Display the result
print("Original Signal:", signal)
print("Kernel:", kernel)
print("Correlated Signal:", correlated_signal)

# Plot the original signal and correlated signal
plt.figure(figsize=(10, 4))

# Plot the original signal
plt.subplot(1, 2, 1)
plt.stem(range(len(signal)), signal, label='Signal')  # Removed `use_line_collection`
plt.title('Original Signal')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Plot the correlated signal
plt.subplot(1, 2, 2)
plt.stem(range(len(correlated_signal)), correlated_signal, label='Correlated Signal', 
         linefmt='g', markerfmt='go')  # Removed `use_line_collection`
plt.title('Correlated Signal')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
