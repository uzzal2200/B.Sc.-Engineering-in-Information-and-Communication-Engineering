```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original signal and its index range
x = np.array([1, 2, 3, 4, 5])  # Original signal values
m = np.arange(-2, 3)           # Index range for the original signal
k = 5                          # Shift amount

# Define the signal shifting function
def signalshift(x, m, k):
    n = m + k  # Shift the indices by k
    y = x      # Signal values remain unchanged
    return y, n
# Call the signal shifting function
y, n = signalshift(x, m, k)

# Plot the original signal
plt.subplot(2, 1, 1)
plt.stem(m, x, markerfmt='bo', basefmt=' ') 
plt.xlabel('m')
plt.ylabel('x(m)')
plt.title('Original Signal')
plt.grid(True)

# Plot the shifted signal
plt.subplot(2, 1, 2)
plt.stem(n, y, markerfmt='ro', basefmt=' ') 
plt.xlabel('n')
plt.ylabel('y(n)')
plt.title('Shifted Signal')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()


```

```python
# Define the signal and its index range
x = np.array([1, 2, 3, 4, 5])  # Example signal
n = np.arange(0, 5)            # Index range of the signal

# Define the sigfold function to compute the folded signal
def sigfold(x, n):
    y = np.flip(x)             # Flip the signal values
    n_folded = -np.flip(n)     # Flip and negate the indices
    return y, n_folded

# Call the sigfold function
y, n_folded = sigfold(x, n)

# Plot the original signal
plt.figure()
plt.subplot(2, 1, 1)
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Original Signal')
plt.grid(True)

# Plot the folded signal
plt.subplot(2, 1, 2)
plt.stem(n_folded, y)
plt.xlabel('n')
plt.ylabel('x(-n)')
plt.title('Folded Signal')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

```python

# Define the signals
x1 = np.arange(1, 6, 1)       # Signal x1: [1, 2, 3, 4, 5]
x2 = np.arange(1, 2.2, 0.2)   # Signal x2: [1, 1.2, 1.4, 1.6, 1.8, 2.0, 2.2]

# Define their index ranges (lengths)
n1 = np.arange(0, len(x1))    # Index range for x1: [0, 1, 2, 3, 4]
n2 = np.arange(0, len(x2))    # Index range for x2: [0, 1, 2, 3, 4, 5, 6]

# Define the sigadd function for signal addition
def sigadd(x1, n1, x2, n2):
    # Determine the unified range of indices
    max_n = max(max(n1), max(n2))
    min_n = min(min(n1), min(n2))
    n = np.arange(min_n, max_n + 1)  # Unified range of indices

    # Initialize arrays with zeros for alignment
    y1 = np.zeros(len(n))
    y2 = np.zeros(len(n))

    # Align x1 to the unified range
    y1[(n >= min(n1)) & (n <= max(n1))] = x1

    # Align x2 to the unified range
    y2[(n >= min(n2)) & (n <= max(n2))] = x2

    # Perform element-wise addition
    y = y1 + y2
    return y, n

# Call the sigadd function to add the signals
y, n = sigadd(x1, n1, x2, n2)

# Plot the first signal x1
plt.subplot(3, 1, 1)
plt.stem(n1, x1)
plt.xlabel('n')
plt.ylabel('x1(n)')
plt.title('Signal x1')
plt.grid(True)

# Plot the second signal x2
plt.subplot(3, 1, 2)
plt.stem(n2, x2)
plt.xlabel('n')
plt.ylabel('x2(n)')
plt.title('Signal x2')
plt.grid(True)

# Plot the addition of the two signals
plt.subplot(3, 1, 3)
plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y(n)')
plt.title('Addition of Two Signals')
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Function for signal multiplication
def sigmult(x1, n1, x2, n2):
    # Determine the unified range for the result
    max_n = max(max(n1), max(n2))
    min_n = min(min(n1), min(n2))
    n = np.arange(min_n, max_n + 1)

    # Initialize arrays with zeros for alignment
    y1 = np.zeros(len(n))
    y2 = np.zeros(len(n))

    # Align x1 to the unified range
    y1[(n >= min(n1)) & (n <= max(n1))] = x1

    # Align x2 to the unified range
    y2[(n >= min(n2)) & (n <= max(n2))] = x2

    # Perform element-wise multiplication
    y = y1 * y2
    return y, n

# Define the signals
x1 = np.arange(1, 6)  # Signal x1 -> [1, 2, 3, 4, 5]
x2 = np.arange(1, 2.2, 0.2)  # Signal x2 -> [1, 1.2, 1.4, ..., 2.2]

# Define their index ranges
n1 = np.arange(len(x1))  # Index range for x1
n2 = np.arange(len(x2))  # Index range for x2

# Multiply the signals using the sigmult function
y, n = sigmult(x1, n1, x2, n2)

# Plot x1
plt.subplot(3, 1, 1)
plt.stem(n1, x1)
plt.xlabel('n')
plt.ylabel('x1(n)')
plt.title('Signal x1')
plt.grid(True)

# Plot x2
plt.subplot(3, 1, 2)
plt.stem(n2, x2)
plt.xlabel('n')
plt.ylabel('x2(n)')
plt.title('Signal x2')
plt.grid(True)

# Plot the product of the signals
plt.subplot(3, 1, 3)
plt.stem(n, y)
plt.xlabel('n')
plt.ylabel('y(n)')
plt.title('Multiplication of Two Signals')
plt.grid(True)

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
