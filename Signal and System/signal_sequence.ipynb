```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to generate the unit step sequence
def stepseq(n0, n1, n2):
    n = np.arange(n1, n2+1)
    x = (n - n0) >= 0
    return x, n

# Generate the step sequence
x, n = stepseq(3, -5, 5)

# Plot the step sequence
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Unit step sequence')
plt.grid(True)
plt.show()
```

```python
# Function to generate the impulse sequence
def impseq(n0, n1, n2):
    n = np.arange(n1, n2+1)  # Generate the range of n
    x = (n - n0) == 0        # Impulse sequence: 1 at n=n0, 0 elsewhere
    return x, n

# Generate the impulse sequence
x, n = impseq(3, -5, 5)

# Plot the impulse sequence
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Unit Sample Space')
plt.grid(True)
plt.show()
```

```python
# Step Function
n = np.arange(-5, 6)
x = (n >= 0)  # Unit Step: 1 for n >= 0, else 0

# Plot
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('u[n]')
plt.title('Unit Step Sequence')
plt.grid(True)
plt.show()
```

```python
# Impulse Function
n = np.arange(-6, 6)
x = (n == 0)  # Unit Impulse: 1 at n=0, else 0

# Plot
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('δ[n]')
plt.title('Unit Impulse Sequence')
plt.grid(True)
plt.show()
```

```python
# Ramp Function
n = np.arange(-5, 6)
x = n * (n >= 0)  # Unit Ramp: n for n >= 0, else 0

# Plot
plt.figure(figsize=(8,3))
plt.stem(n, x)
plt.xlabel('n')
plt.ylabel('r[n]')
plt.title('Unit Ramp Sequence')
plt.grid(True)
plt.show()
```
