```python
import numpy as np
import matplotlib.pyplot as plt

def step_impulse(n, shift):
    return np.where(n == shift, 1, 0)


# n1 = np.arange(-10, 11)

n1 = np.array([-1,-2,-3, -5, 0, 1, 2, 3,4,5])

# np.where() হলো NumPy লাইব্রেরির একটি ফাংশন, যা শর্তের উপর ভিত্তি করে অ্যারে থেকে মান বের করে বা নতুন অ্যারে তৈরি করে।
# এটি মূলত If-Else এর মতো কাজ করে
# k>0 হলে ডানদিকে শিফট হবে।
# k<0 হলে বামদিকে শিফট হবে।
shift_value = 2  # এখানে শিফট মান পরিবর্তন করতে পারেন
impulse_signal = step_impulse(n1, shift_value)

plt.stem(n1, impulse_signal)
plt.title('Shifted Impulse Function')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()
```

```python
def unit_step(n):

    return np.where(n >= 0, 1, 0)

step_signal =  unit_step(n1)

plt.stem(n1, step_signal)

plt.title(' Unit Function')
plt.xlabel('n')
plt.ylabel('u(n)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# t = নিরবচ্ছিন্ন সময়ের জন্য অ্যারে তৈরি
t = np.linspace(-10, 10, 100)  # -10 থেকে 10 পর্যন্ত 1000টি পয়েন্ট

# ইউনিট স্টেপ ফাংশন
def unit_step(t):
    return np.where(t >= 0, 1, 0)

# ইউনিট স্টেপ সিগনাল
step_signal = unit_step(t)

# গ্রাফ অঙ্কন
plt.plot(t, step_signal, label='u(t)', color='blue')
plt.title('Continuous Unit Step Function')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.grid(True)
plt.axvline(x=0, color='gray', linestyle='--')  # y-axis দেখানোর জন্য
plt.axhline(y=0, color='gray', linestyle='--')  # x-axis দেখানোর জন্য
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions
def impulse_signal(n):
    return np.where(n == 0, 1, 0)  # δ[n] = 1 for n=0, else 0

def step_signal(n):
    return np.where(n >= 0, 1, 0)  # u[n] = 1 for n >= 0, else 0

def ramp_signal(n):
    return np.where(n >= 0, n, 0)  # r[n] = n for n >= 0, else 0

# Generate signals
n1 = np.arange(-10, 11)  # For impulse and step
n2 = np.arange(-5, 6)    # For ramp

impulse = impulse_signal(n1)
step = step_signal(n1)
ramp = ramp_signal(n2)

# Create a figure with 3 subplots
plt.figure(figsize=(10, 6))

# Plot Unit Impulse Function
plt.subplot(3, 1, 1)
plt.stem(n1, impulse)
plt.xlabel('n')
plt.ylabel('δ[n]')
plt.title('Unit Impulse Function')
plt.grid(True)

# Plot Unit Step Function
plt.subplot(3, 1, 2)
plt.stem(n1, step)
plt.xlabel('n')
plt.ylabel('u[n]')
plt.title('Unit Step Function ')
plt.grid(True)

# Plot Unit Ramp Function
plt.subplot(3, 1, 3)
plt.stem(n2, ramp)
plt.xlabel('n')
plt.ylabel('r[n]')
plt.title('Ramp Function')
plt.grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
